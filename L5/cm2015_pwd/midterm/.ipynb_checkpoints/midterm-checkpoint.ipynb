```python
import pandas as pd
import glob
from dateutil.parser import parse

matches = []
for infile in glob.glob("data/*.csv"):
    data = pd.read_csv(infile)
    matches.append(data)
matches = pd.concat(matches, ignore_index=True)

matches.head()
```

```python
matches = matches[['Date', 'HomeTeam','AwayTeam','FTHG','FTAG','B365H','B365D','B365A']].copy()

matches
```

```python
matches['Date'] = pd.to_datetime(matches['Date'], dayfirst=True)

matches.sort_values(by='Date', axis=0, ascending=True, ignore_index=True, inplace=True)

matches
```

```python
def determineActualWinner():
	for i, match in matches.iterrows():
		if match['FTHG'] > match['FTAG']:
			return match['HomeTeam']
		elif match['FTHG'] < match['FTAG']:
			return match['AwayTeam']
		else:
			return "DRAW"

matches['Actual Winner'] = determineActualWinner()

# matches['Actual Winner'] = [match['HomeTeam'] if match['FTHG'] > match['FTAG'] else match['AwayTeam'] if match['FTHG'] < match['FTAG'] else "DRAW" for i, match in matches.iterrows()]

# matches['Predicted Winner'] = [match['HomeTeam'] if match['B365H'] < min(match['B365A'],match['B365D']) else match['AwayTeam'] if match['B365A'] < min(match['B365H'],match['B365D']) el#se "DRAW" for i, match in matches.iterrows()]

matches
```
